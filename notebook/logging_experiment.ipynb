```python
import logging
import os
from datetime import datetime
```

```python
logs_dir=os.path.join(os.getcwd(),"logs")
```

```python
os.makedirs("logs", exist_ok=True)
```

```python
LOG_FILE = f"{datetime.now().strftime('%m_%d_%Y_%H_%M_%S')}.log"
```

```python
#  This is the log file (name of the log file would be current date and time)
LOG_FILE
```

## with Todays date log file will be creates

## If I EXECUE MY CODE TOMMOROW THAN WITH TOMMOROW DATE MY LOG FILE WILLBE STORED

```python
LOG_FILE_PATH = os.path.join(logs_dir,LOG_FILE)
```

```python
# basic cofig of logging asc time means current time  level in info , and name of the log file  , line is the we have to capture the log at this line and , message in the end
logging.basicConfig(
    filename=LOG_FILE_PATH,
    format = "[ %(asctime)s ] %(levelname)s %(name)s (line:5(lineno)d) - %(message)s",
    level=logging.INFO,
)
```

```python
logger=logging.getLogger("DocumentPortal")
```
