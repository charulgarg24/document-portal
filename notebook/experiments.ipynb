Key Differences


Splitter Module	Split Method	Pros	Cons	Use Case

CharacterTextSplitter	Fixed characters / separator	Fast, simple	Cuts meaning, may break sentences	Quick splits, small texts

RecursiveCharacterTextSplitter	Recursive separators	Preserves semantic meaning	Slower, complex	RAG pipelines, document QA

TokenTextSplitter	LLM tokens	Exact token control, fits model limits	Needs tokenizer	Embeddings, GPT input preparation

SpacyTextSplitter / SentenceTextSplitter	Sentence segmentation (NLP)	Keeps sentences intact	Chunk size varies, dependency	Summarization, QA, context-sensitive tasks

Chunking using recursivve text splitter

This is a experimetn thing theere is no specific way to or size to do chunking

```python
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=150,
    length_function=len
)
```

```python
doc=text_splitter.split_documents(documents)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

# Suppose doc[12].page_content is the text content of your document
text = doc[12].page_content

#  Correct: use encode
vector = model.encode(text)

print(len(vector))       # typically 384 dimensions for MiniLM
print(vector[:10])       # just showing first 10 numbers
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import SentenceTransformerEmbeddings

# Wrap your local model for LangChain
embedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Build FAISS vector store locally
vector_store = FAISS.from_documents(doc, embedding_model)
```

## This is a retrival process
means from the vector database wew are going to fetch or retrive or rank the most appropriate k result

```python
relevant_doc = vector_store.similarity_search("what is kinetic energy?",k=10)
```

```python
relevant_doc
```

```python
retriever=vector_store.as_retriever()
```

```python
retriever.invoke("physics kinetic energy experiemnt.")
```

## contect i am retrieving from vector datatabse
## Context: based on the question retrieving the info from the vector database

```python
prompt_template = """
        Answer the question based on the context provided below. 
        If the context does not contain sufficient information, respond with: 
        "I do not have enough information about this."

        Context: {context}

        Question: {question}

        Answer:"""
```

```python
from langchain.prompts import PromptTemplate
```

```python
prompt=PromptTemplate(
    template=prompt_template,
    input_variable=["context","question"]
)
```

```python
prompt
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
parser=StrOutputParser()
```

```python

def format_docs(docs):
    return "\n\n".join([doc.page_content for doc in docs])
```

```python

from langchain_core.runnables import RunnablePassthrough
```

```python

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
rag_chain.invoke("tell  me about the kinetic enrgy in physics?")
```

```python
rag_chain.invoke("can you tell me what is parallel axis energy ?") 
```
